Learning how to use Python's Mersenne Twister and plotlib

```python
import random 
import matplotlib.pyplot as plt
random.seed(5734985792340572345) #The current system time is used by default
random.random()
```

```python
random.randrange(0,50)
```

```python
#Random permutation of a list
a = list(range(0,50))
plt.bar(a, a)
random.shuffle(a)
print(a)
plt.bar(a, height=list(range(0,50)))
```

```python
random.sample(a, 10)
```

```python
l = [random.expovariate(42) for i in range(0,5000)]
plt.hist(l, bins=100)
plt.show()
```

```python
l = [random.gauss(42,42) for i in range(0,5000)]
plt.hist(l, bins=100)
plt.show()
```
