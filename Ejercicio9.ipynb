```python
import random as rdm
import matplotlib.pyplot as plt
import math
```

```python
#Ejercicio 9

def gen_x():
    u = rdm.random()
    i = 1
    if u <= 0.5:
        p = 0.5
        F = p
        while u > F:
            p *= 0.5
            F += p
            i += 1
    else:
        p = 1/3
        F = p
        while u > F:
            p *= 2/3
            F += p
            i += 1
    return i

        
```

```python
l = [gen_x() for i in range(500)]
plt.hist(l, bins=15)
plt.show()
```
