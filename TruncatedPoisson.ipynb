```python
import numpy as np 
import matplotlib.pyplot as plt
import math
```

```python
np.random.poisson()
plt.plot(np.random.poisson())
plt.show()
```

```python
def truncated_poisson(l, k, i):
    if 0<= i <= k:
        poisson = lambda l, i: l**i / math.factorial(i) * math.exp(-l)
        numerator = poisson(l, i)
        denominator = sum(poisson(l, j) for j in range(k+1))
        return numerator / denominator
    else:
        return 0
```

```python
truncated_poisson(1, 10, 10)
plt.plot(range(20), [truncated_poisson(5,10,j) for j in range(20)])
plt.show()
```
